## 1.2 Extract Image Data into List Structure

```python
from PIL import Image
import os

# Path to dataset
dataset_path = r"C:\Users\shash\.cache\kagglehub\datasets\ubitquitin\geolocation-geoguessr-images-50k\versions\1\compressed_dataset"

images = []
labels = []

# Loop through country folders
for country_folder in os.listdir(dataset_path):
    country_path = os.path.join(dataset_path, country_folder)
    if os.path.isdir(country_path):
        for img_file in os.listdir(country_path):
            if img_file.endswith('.jpg'):  # Or whatever image format it is
                img_path = os.path.join(country_path, img_file)
                img = Image.open(img_path)
                img = img.resize((224, 224))  # Resize image
                images.append(img)  # Add image to list
                labels.append(country_folder)  # Add label to list
```

## 1.3 Convert Lables to Numeric

```python
from sklearn.preprocessing import LabelEncoder

# Encode labels
label_encoder = LabelEncoder()
lables_encoded = label_encoder.fit_transform(labels)

print(lables_encoded[:10])
```
